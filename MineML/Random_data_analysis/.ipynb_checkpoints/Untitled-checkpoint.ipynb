```python
import pandas as pd 
import numpy as np
import tensorflow as tf
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
import seaborn as sns
import matplotlib.pyplot as plt
import category_encoders as ce
%matplotlib inline
```

```python
df_train = pd.read_csv("train.csv")
df_test = pd.read_csv("test.csv")
df_sample_submission = pd.read_csv("sample_submission.csv")
```

```python
label_encoder = LabelEncoder()
df_train['X11'] = label_encoder.fit_transform(df_train['X11'])
df_train['X13'] = label_encoder.fit_transform(df_train['X13'])
df_train['X17'] = label_encoder.fit_transform(df_train['X17'])

df_train = df_train.drop(['X1','X3','X6','X14','X15','X16','X18'], axis = 1)
```

```python
df_train.corr()
```

```python
target_encoder = ce.TargetEncoder()
df_train['X11'] = target_encoder.fit_transform(df_train['X11'],df_train['y'])
df_train['X13'] = target_encoder.fit_transform(df_train['X13'],df_train['y'])
df_train['X17'] = target_encoder.fit_transform(df_train['X17'],df_train['y'])
```

```python
df_missng_x5 = df_missng_x4 = df_train[df_train['X4'].isna() & df_train['X5'].isna()]
df_not_missing_x5 = df_not_missing_x4 = df_train[df_train['X4'].notna() & df_train['X5'].notna()]
columns_for_X4 = ['y','X9','X10']
columns_for_X5 = ['y','X7','X8','X9','X10','X12','X13','X17']
y_x4 = df_not_missing_x4['X4']
df_not_missing_x4_columns =  df_not_missing_x4[columns_for_X4]
y_x5 = df_not_missing_x5['X5']
df_not_missing_x5_columns =  df_not_missing_x5[columns_for_X5]
```

```python
rfr = RandomForestRegressor()
rfr.fit(df_not_missing_x4_columns,y_x4)
predictions_x4 = rfr.predict(df_missng_x4[columns_for_X4])
rfr = RandomForestRegressor()
rfr.fit(df_not_missing_x5_columns,y_x5)
predictions_x5 = rfr.predict(df_missng_x5[columns_for_X5])
```

```python
missing_indexes = df_missng_x4.index
i = 0
for index in missing_indexes:
    df_train.loc[index, 'X4'] = predictions_x4[i]
    df_train.loc[index, 'X5'] = predictions_x5[i]
    i = i+1
df_train.isna().sum()
```
