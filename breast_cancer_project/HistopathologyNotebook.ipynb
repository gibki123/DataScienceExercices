```python
def processImage(image_to_process):
    resized_image = cv2.resize(image_to_process, (WIDTH,HEIGHT), interpolation=cv2.INTER_CUBIC)
    img_g = cv2.cvtColor(resized_image, cv2.COLOR_BGR2GRAY)
    grayg = cv2.GaussianBlur(img_g, (5, 5), 0)
    
    return grayg
```

```python
x = []
y = []

df = pd.DataFrame

for directory in image_directories:
    if directory.endswith("0.png"):
        image = cv2.imread(directory)
        processed_image = processImage(image)
        x.append(processed_image)
        y.append(0)
    else:
        image = cv2.imread(directory)
        processed_image = processImage(image)
        x.append(processed_image)
        y.append(1)
    
df['Images'] = x
df['Labels'] = y
    
```
